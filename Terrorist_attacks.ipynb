```python
year2013 = Data[Data['Attack Year']==2013]
year2013
```

# Comparision between India, Pakistan and Israel

**Data of India**

```python
df_India = Data[Data['Country Name']=="India"]
df_India
```

**Data of Pakistan**

```python
df_Pak = Data[Data['Country Name']=="Pakistan"]
df_Pak
```

**Data of Isreal**

```python
df_Israel = Data[Data['Country Name']=="Israel"]
df_Israel
```

Now compare the data of **three countries India, Pakistan and Israel** from 2012 to 2021.

```python
# set width of bar
barwidth = 0.25
fig = plt.subplots(figsize=(10,6))

# set height of bar
India = df_India['No. of Attacks'].tail(10)
Pak = df_Pak['No. of Attacks'].tail(10)
Israel = df_Israel['No. of Attacks'].tail(10)

# set position of bar on X axis
br1 = np.arange(len(India))
br2 = [x + barwidth for x in br1]
br3 = [x + barwidth for x in br2]

# Make the plot
plt.bar(br1,India,color='red',width=barwidth,edgecolor='black',label='India')
plt.bar(br2,Pak,color='green',width=barwidth,edgecolor='black',label='Pakistan')
plt.bar(br3,Israel,color='blue',width=barwidth,edgecolor='black',label='Israel')

# adding Xticks
plt.title('Number of terrorist attack in India, Pakistan ans Isreal')
plt.xlabel('Attack Year',fontweight='bold',fontsize=15)
plt.ylabel('Number of attacks',fontweight='bold',fontsize=15)
plt.xticks([r+barwidth for r in range(len(India))],[2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])
plt.legend()
plt.show()
```
